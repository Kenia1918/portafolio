### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Calculamos el ingreso total por cada tienda.

ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f'El ingreso de la tienda 1 es de: {ingreso_tienda1}')

print(f'El ingreso de la tienda 2 es de: {ingreso_tienda2}')

print(f'El ingreso de la tienda 3 es de: {ingreso_tienda3}')

print(f'El ingreso de la tienda 4 es de: {ingreso_tienda4}')

#Creamos un diccionario para almacenar los ingresos de cada tienda.

ingreso_total = {'Tienda1': ingreso_tienda1, 'Tienda2': ingreso_tienda2, 'Tienda3': ingreso_tienda3, 'Tienda4': ingreso_tienda4}

#Verificamos que tienda es la mejor y cual es la peor segun sus ingresos.
mejor_tienda = max(ingreso_total, key=ingreso_total.get)
peor_tienda = min(ingreso_total, key=ingreso_total.get)

print(f'La mejor tienda es {mejor_tienda} con un ingreso de {ingreso_total[mejor_tienda]}')
print(f'La peor tienda es {peor_tienda} con un ingreso de {ingreso_total[peor_tienda]}')

#Creamos un gráfico para mostrar las ventas de cada tienda.

#Creamos las listas con los datos que necesitamos para crear el gráfico.

tiendas = list(ingreso_total.keys())
ingresos = list(ingreso_total.values())

#Creamos el gráfico.

import matplotlib.pyplot as plt

plt.bar(tiendas, ingresos, color='green')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.title('Ingresos por tienda')
plt.show()
```

# 2. Ventas por categoría

```python
#Contamos cuantos productos por categoria se venden en cada tienda

categorias_tienda1 = tienda['Categoría del Producto'].value_counts()
categorias_tienda2 = tienda2['Categoría del Producto'].value_counts()
categorias_tienda3 = tienda3['Categoría del Producto'].value_counts()
categorias_tienda4 = tienda4['Categoría del Producto'].value_counts()

print(f'Cantidad de productos por categoría en la tienda 1: \n{categorias_tienda1}')
print(f'\nCantidad de productos por categoría en la tienda 2: \n{categorias_tienda2}')
print(f'\nCantidad de productos por categoría en la tienda 3: \n{categorias_tienda3}')
print(f'\nCantidad de productos por categoría en la tienda 4: \n{categorias_tienda4}')
```

```python
#Creamos un nuevo DataFrame con las categorias y todas las tiendas.

cat_tiendas = pd.concat([categorias_tienda1, categorias_tienda2, categorias_tienda3, categorias_tienda4], axis=1)
cat_tiendas.columns = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']
cat_tiendas
```

```python
#Creamos el grafico para comparar las categorias de todas las tiendas.
#Ocuparemos un gráfico de barras donde se visualizaran 4 barras por categorías, cada barra indica una tienda, y se muestra que productos de que categoría
#se venden más y que categoría se vende menos.
import matplotlib.pyplot as plt

cat_tiendas.plot(kind='bar', figsize=(12, 10))
plt.title('Comparación por categoria de las 4 tiendas.')
plt.xlabel('Categorías del Producto')
plt.ylabel('Cantidad de productos')
plt.show()

```

# 3. Calificación promedio de la tienda


```python
#Calculamos el promedio de calificación de todas las tiendas.
#Utilizamos .mean() para obtener la media de cada tienda.

promedio_tienda1 = round(tienda['Calificación'].mean(),2)
promedio_tienda2 = round(tienda2['Calificación'].mean(),2)
promedio_tienda3 = round(tienda3['Calificación'].mean(),2)
promedio_tienda4 = round(tienda4['Calificación'].mean(),2)

print(f'El promedio de calificación de la tienda 1 es de: {promedio_tienda1}')
print(f'El promedio de calificacion de la tienda 2 es de: {promedio_tienda2}')
print(f'El promedio de calificacion de la tienda 3 es de: {promedio_tienda3}')
print(f'El promedio de calificación de la tienda 4 es de: {promedio_tienda4}')
```

```python
#Realizamos el gráfico que muestra las medias de las evaluaciones de cada tienda.

import matplotlib.pyplot as plt

#Hacemos las listas que ocuparemos para el gráfico.

tiendas = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']
promedios = [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]

#Creamos el gráfico.

plt.bar(tiendas, promedios, color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('Promedio de calificaciones')
plt.title('Promedio de calificaciones por tienda')

plt.show()
```

# 4. Productos más y menos vendidos

```python
#Buuscaremos los productos más y menos vendidos por cada tienda.

productos_tienda1 = tienda['Producto'].value_counts()
productos_tienda2 = tienda2['Producto'].value_counts()
productos_tienda3 = tienda3['Producto'].value_counts()
productos_tienda4 = tienda4['Producto'].value_counts()

#Imprimos todos los productos vendidos por cada tienda.
print(f'Productos vendidos en la tienda 1: \n{productos_tienda1}')
print(f'\nProductos vendidos en la tienda 2: \n{productos_tienda2}')
print(f'\nProductos vendidos en la tienda 3: \n{productos_tienda3}')
print(f'\nProductos vendidos en la tienda 4: \n{productos_tienda4}')

#Mostramos el  producto más y menos vendido por tienda.
print(f'\nEl producto más vendido en la tienda 1 es: \n{productos_tienda1.head(1)}')
print(f'\nEl producto menos vendido en la tienda 1 es: \n{productos_tienda1.tail(1)}')

print(f'\nEl producto más vendido en la tienda 2 es: \n{productos_tienda2.head(1)}')
print(f'\nEl producto menos vendido en la tien  da 2 es: \n{productos_tienda2.tail(1)}')

print(f'\nEl producto más vendido en la tienda 3 es: \n{productos_tienda3.head(1)}')
print(f'\nEl producto menos vendido en la tienda 3 es: \n{productos_tienda3.tail(1)}')

print(f'\nEl producto más vendido en la tienda 4 es: \n{productos_tienda4.head(1)}')
print(f'\nEl producto menos vendido en la tienda 4 es: \n{productos_tienda4.tail(1)}')
```

```python

#Creamos el gráfico que muestra los productos más y menos vendidos por tienda.

import matplotlib.pyplot as plt

#Gráfico de los productos más y menos vendidos en la tienda 1.
figure, axis = plt.subplots(2, 2, figsize=(10, 10))
axis[0, 0].bar(productos_tienda1.head(1).index, productos_tienda1.head(1).values, color='green', label='Mas vendido')
axis[0, 0].bar(productos_tienda1.tail(1).index, productos_tienda1.tail(1).values, color='red', label='Menos vendido')
axis[0, 0].set_xlabel('Productos')
axis[0, 0].set_ylabel('Cantidad de productos')
axis[0, 0].set_title('Productos más y menos vendidos en la tienda 1')
axis[0, 0].legend() #Se añade la leyenda para identificar las barras


#Gráfico de los productos más y menos vendidos en la tienda 2.
axis[0,1].bar(productos_tienda2.head(1).index, productos_tienda2.head(1).values, color='green', label='Mas vendido')
axis[0,1].bar(productos_tienda2.tail(1).index, productos_tienda2.tail(1).values, color='red', label='Menos vendido')
axis[0,1].set_xlabel('Productos')
axis[0,1].set_ylabel('Cantidad de productos')
axis[0,1].set_title('Productos más y menos vendidos en la tienda 2')
axis[0,1].legend() #Se añade la leyenda para identificar las barras

#Gráfico de los productos más y menos vendidos en la tienda 3.
axis[1,0].bar(productos_tienda3.head(1).index, productos_tienda3.head(1).values, color='green', label='Mas vendido')
axis[1,0].bar(productos_tienda3.tail(1).index, productos_tienda3.tail(1).values, color='red', label='Menos vendido')
axis[1,0].set_xlabel('Productos')
axis[1,0].set_ylabel('Cantidad de productos')
axis[1,0].set_title('Productos más y menos vendidos en la tienda 3')
axis[1,0].legend()

#Gráfico de los productos más y menos vendidos en la tienda 4.
axis[1,1].bar(productos_tienda4.head(1).index, productos_tienda4.head(1).values, color='green', label='Mas vendido')
axis[1,1].bar(productos_tienda4.tail(1).index, productos_tienda4.tail(1).values, color='red', label='Menos vendido')
axis[1,1].set_xlabel('Productos')
axis[1,1].set_ylabel('Cantidad de productos')
axis[1,1].set_title('Productos más y menos vendidos en la tienda 4')
axis[1,1].legend()

plt.show()


```

# 5. Envío promedio por tienda

```python
#Se calcula el gasto promedio de los envios.

envio_promedio_tienda1 = round(tienda['Costo de envío'].mean(),2)
envio_promedio_tienda2 = round(tienda2['Costo de envío'].mean(),2)
envio_promedio_tienda3 = round(tienda3['Costo de envío'].mean(),2)
envio_promedio_tienda4 = round(tienda4['Costo de envío'].mean(),2)

print(f'El envío promedio de la tienda 1 es: {envio_promedio_tienda1}')
print(f'El envío promedio de la tienda 2 es: {envio_promedio_tienda2}')
print(f'El envío promedio de la tienda 3 es: {envio_promedio_tienda3}')
print(f'El envío promedio de la tienda 4 es: {envio_promedio_tienda4}')

#Creamos un DataFrame de los pomedios de envios.

envios_promedio = {'Tienda1': envio_promedio_tienda1, 'Tienda2': envio_promedio_tienda2, 'Tienda3': envio_promedio_tienda3, 'Tienda4': envio_promedio_tienda4}
envios_promedio
```

```python
#Graficamos los costos promedio de envio de las tiendas.

import matplotlib.pyplot as plt

tiendas = list(envios_promedio.keys())
envios = list(envios_promedio.values())

plt.bar(tiendas, envios, color='orange')
plt.title('Envios promedio por tienda')
plt.xlabel('Tiendas')

for i, value in enumerate(envios):
    plt.text(i, value + 5,  str(value), ha='center')
plt.show()
```
